```python
df = pd.read_csv('/content/drive/MyDrive/Text/Tweets_sentiment_16M.csv')
df
```

```python
filtered_df = df[df['target'] == 2] #เช็ตเจ๋ยๆว่ามันไม่มี2จริงมั้ย
filtered_df
```

```python
df['score'] = df['sentiment_score'].replace([1, 2, 3], 0) #แทนที่การทำนายค่าของเซนติเม้นของคะแนน 1,2,3 เป็น 0
#เนื่องจากไม่มี2
df['score'] = df['score'].replace([4, 5], 4) #แทนที่การทำนายค่าของเซนติเม้นของคะแนน 4,5 เป็น 4
df
```

```python
filtered_df = df[df['score'].isin([1, 3, 5])] #แค่ดูว่ามันถูกแทนที่ไปจริงมั้ย
filtered_df
```

```python
matches = (df['target'] == df['score']).sum() #เปรียบเทียบค่าในcolum target and score
total_rows = len(df)
percentage = (matches / total_rows) * 100 #เอามาหาค่าเฉลี่ยความต่างกันของข้อมูล
percentage
```

```python
different_values = df.loc[df['target'] != df['score'], ['text2', 'target', 'score']] #เอาค่าที่มันต่างกันออกมาให้ดู
different_values
```
