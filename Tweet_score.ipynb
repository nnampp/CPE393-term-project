```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Tweets_sentiment.csv')
df.sample(5)
```

```python
conditions = [
    (df['sentiment_score'] == 1) | (df['sentiment_score'] == 2 | (df['sentiment_score'] == 3)),
    (df['sentiment_score'] == 4) | (df['sentiment_score'] == 5)
]
```

```python
letters = ['0', '4']
```

```python
df['score'] = np.select(conditions, letters)
df
```

```python
df.to_csv('Tweets_sentiment_convert_score_new.csv', index=False)
```
